# Que significan los partidos que terminan "en goleada"

La mayoría de los partidos del superbowl son muy parejos. Por qué? Es el partido final de la Liga de fútbol americano, los equipos tienen que pasar por varias rondas antes de llegar a la final, como la temporada regular, los playoffs, etc.

Pero que pasa con los partidos que son una goleada por parte de un equipo al otro? Terminan siendo aburridos? La gente los ve completos? O empiezan a perder espectadores? 

Vamos a ver ahora el dataset de TV, para ver si esto de verdad sucede

```python
#Vamos a combinar los datos de TV y los datos de Superbowl para ver si esto tiene relación
#Primero, vemos que el superbowl fue transmitido en dos cadenas en su primera emisión, así que vamos a combinar los datos
games_tv = pd.merge(tv[tv['super_bowl']>1], super_bowls, on='super_bowl')

import seaborn as sns

sns.regplot(x='difference_pts', y='share_household', data=games_tv)
```

# La industria del advertising en el tiempo

Podemos apreciar, con la línea de regresión, que dejar de ver el partido, es común si la diferencia de puntos es muy alta, mientras que si la diferencia de puntos es menor, la audiencia lo sigue viendo

Pero independientemente de esto, lo que más llama la atención del super bowl es el show de medio tiempo. 30 segundos de anuncios cuestan aproximadamente 5 millones de dolares, pero siempre fue así? 

```python
# Creamos una figura subplot de 3 x 1, y en la primera, activamos el primer subplot

#Costo de los anuncios en el tiempo
plt.subplot(3,1,1)
plt.plot(tv.super_bowl, tv.ad_cost, color='#648FFF')
plt.title('Cost over time')

#Activamos el segundo subplot
#Cantidad de visualizaciones conforme los superbowls
plt.subplot(3,1,2)
plt.plot(tv.super_bowl, tv.avg_us_viewers, color='#DC267F')
plt.title('Viewers over time')

#Activamos el tercer subplot
#Calificación del hogar
plt.subplot(3,1,3)
plt.plot(tv.super_bowl, tv.rating_household, color="#FFB000")
plt.title('Household rating')

#Mejorar el espacio entre las gráficas
plt.tight_layout()
```

# Quien tiene más apariciones en show de medio tiempo?

Vamos a ver a los músicos que se presentaron más de una vez en un show de medio tiempo

```python
#Contamos las apariciones en espectáculos de medio tiempo de cada músico y ordenamos de mayor a menor
halftime_appearances = halftime_musicians.groupby('musician').count()['super_bowl'].reset_index()
halftime_appearances = halftime_appearances.sort_values('super_bowl', ascending=False)

halftime_appearances[halftime_appearances['super_bowl']>1]
```

# Quien tocó más canciones?

Primero vamos a quitar las bandas, ya que aparecen muchos nombres con la palabra "Marching" y "Spirit"
Luego filtraremos por Super Bowls después del Super Bowl XX para abordar el problema de los datos faltantes, luego vamos a ver quién tiene la mayor cantidad de canciones.

```python
#Filtrando bandas con Marching
no_bands = halftime_musicians[~halftime_musicians.musician.str.contains("Marching")]
no_bands = no_bands[~no_bands.musician.str.contains('Spirit')]

#Graficamos un histograma con el número de canciones por show
most_songs = int(max(no_bands['num_songs'].values))
plt.hist(no_bands.num_songs.dropna(), bins=most_songs)
plt.xlabel('Number of Songs Per Halftime Show Performance')
plt.ylabel('Number of Musicians')
plt.show()

# Sort the non-band musicians by number of songs per appearance...
no_bands = no_bands.sort_values('num_songs', ascending=False)
# ...and display the top 15
display(no_bands.head(15))
```
